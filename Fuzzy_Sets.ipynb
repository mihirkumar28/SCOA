```python
from matplotlib import pyplot as plt
```

```python
a = {"x1": 0.3, "x2": 0.5, "x3":0.12, "x4":0.73}
plt.plot([a[e] for e in a],color='r')
plt.xticks([i for i in range(len(a))],[e for e in a])
plt.show()
```

```python
b = {"x1": 0.8, "x2": 0.2, "x3":1.0, "x4":0.45}
plt.plot([b[e] for e in b],color='b')
plt.xticks([i for i in range(len(b))],[e for e in b])
plt.show()
c = {}
```

```python
for i in a:
    if a[i]>b[i]:
        c[i]=a[i]
    else:
        c[i]=b[i]

print("Union of Set A and B",c)
plt.plot([a[e] for e in a],color='r')
plt.plot([b[e] for e in b],color='b')
plt.plot([c[e] for e in c],linewidth=5,color='y')
plt.xticks([i for i in range(len(b))],[e for e in b])
plt.show()
```

```python
d={}
for i in a:
    if a[i]<b[i]:
        d[i]=a[i]
    else:
        d[i]=b[i]

print("Intersection Set A and B",d)
plt.plot([a[e] for e in a],color='r')
plt.plot([b[e] for e in b],color='b')
plt.plot([c[e] for e in c],linewidth=5,color='y')
plt.xticks([i for i in range(len(b))],[e for e in b])
plt.show()
```

```python
k={}
for i in a:
    k[i]=round(1-a[i],2)
print("Complemet of Set A :" ,k)

plt.plot([a[e] for e in a],color='r')
plt.plot([k[e] for e in k],color='g')
plt.xticks([i for i in range(len(b))],[e for e in b])
plt.show()

for i in b:
    k[i]=round(1-b[i],2)
print("Complemet of Set B :" ,k) 

plt.plot([b[e] for e in b],color='r')
plt.plot([k[e] for e in k],color='g')
plt.xticks([i for i in range(len(b))],[e for e in b])
plt.show()
```

```python
d={}
for i in a:
    d[i]=round((min(a[i],1-b[i])),2)    
print(f"Set A: {a} \nSet B: {b}","\nDifference A-B:" ,d)    

plt.plot([a[e] for e in a],color='r')
plt.plot([b[e] for e in b],color='b')
plt.plot([d[e] for e in d],linewidth=5,color='y')
plt.xticks([i for i in range(len(a))],[e for e in a])
plt.show()
```

```python
#CARTESIAN PRODUCT
R = [[] for i in range(len(a))]
i = 0
for x in a:
    for y in b:
        R[i].append(min(a[x], b[y]))
    i += 1
print("R :\n",R,"\n") 

S = [[] for i in range(len(a))]
i = 0
for x in a:
    for y in b:
        S[i].append(round(min(1-a[x], 1-b[y]),2))
    i += 1
print("S :\n",S) 
```

```python
import numpy as np
R=np.array(R)
S=np.array(S)
tmp = np.zeros((R.shape[0], S.shape[1]))
t = list()
for i in range(len(R)):
    for j in range(len(S[0])):
        for k in range(len(S)):
            t.append(round(min(R[i][k], S[k][j]), 2))
        tmp[i][j] = max(t)
        t.clear()
print(tmp)        
```
